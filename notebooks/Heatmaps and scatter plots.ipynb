```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Given data sorted along age

```python
filepath = "./heart.csv"
heart_df = pd.read_csv(filepath)
heart_df.dropna(inplace = True)
heart_df.sort_values(by = "age", inplace = True)
display(heart_df)
```


# Heatmap

```python
# display(heart_df.corr())
ax = sns.heatmap(heart_df.corr())
```

# Scatter plots for different parameters and their relation with age

```python
print(heart_df.columns)
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(
    data=heart_df, x="age", y="cp", 
    legend="full"  
)
```

```python
sns.scatterplot(
    data=heart_df, x="age", y="trestbps", 
    legend="full" 
)
```

```python
fig, axes = plt.subplots(4, 3, figsize = (15, 15))
plt.subplots_adjust(
    hspace = 0.5
)
for ax, col_name in zip(axes.flatten(), heart_df.columns.values[2:]):
    sns.scatterplot(data = heart_df, x = "age", y = col_name, ax = ax)
    ax.set_title(col_name)
```
